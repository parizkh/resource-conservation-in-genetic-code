# Structural principles of resource conservation in the genetic code

This notebook performs the analysis of the structural principles of resource conservation in the genetic code, in particular:

1. The relationship between number of codons encoding nitrogen-rich amino acids and ERMC in nitrogen in codes generated by amino acid permutation.
2. The relationship between the 'line', 'square' and 'diagonal' arrangements of codes generated by quartet shuffling and ERMC in nitrogen.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
def read_code(fileName):
    code = {}
    with open(fileName, 'r') as f:
        lines = f.read().split("\n")
        # remove the header
        lines.pop(0)
        for l in lines:
            if l!="":
                splitLine = l.split("\t")
                #print(splitLine)
                code[splitLine[1]] = splitLine[0]
    return code
```

## 1. In codes generated by amino acid permutation, ERMC in nitrogen is correlated with the number of codons of nitrogen-rich amino acids

Function to generate codes by amino acid permutation:

```python
# standard_code ... dictionary containing the standard code; keys = codons, values = amino acids
def rand_aa_permutation(standard_code, seed = 0):
    np.random.seed(seed)
    # all amino acids
    aas = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', '*']
    # dictionary to store the code
    code = {}
    # randomly shuffle the amino acids 
    shuffled_aas = np.random.permutation(20)
    # the last one is always stop
    shuffled_aas = np.append(shuffled_aas, [20])
    # go over the standard code and always change amino acid i to shuffled_aas[i]
    for codon, aa in standard_code.items():
        code[codon] = aas[shuffled_aas[aas.index(aa)]]
    return code
```

Function to count number of codons of nitrogen-rich amino acids:

```python
def number_N_rich_codons(code):
    n = 0
    for codon, aa in code.items():
        if aa in ["H", "K", "N", "Q", "R", "W"]:
            n += 1
    return n
```

Functions to compute ERMC:

```python
def gen_neighbouring_codons(c):
    res = []
    for i in [0,1,2]:
        for b in ['A', 'C', 'G', 'T']:
            if c[i]!=b:
                res.append(c[:i] + b + c[(i+1):])
    return res

def get_mutation_type(old, new):
    for i in range(3):
        if old[i]!=new[i]:
            return old[i]+new[i]

def ERMC(code, property_dict, mutRates, codons_frequencies, abs_val=False):
    ermc = 0
    
    codons = list(code.keys())

    for c in codons:
        aa = code[c]
        freq = codons_frequencies[c]
        if aa!="*":
            neighs = gen_neighbouring_codons(c)
            for n in neighs:
                n_aa = code[n]
                if n_aa!="*":
                    # mutation type
                    mutType = get_mutation_type(c,n)
                    # use absolute values in the computation of ERMC (used for mol. volume, PR, hydropathy)
                    if abs_val == True:
                        ermc += freq*mutRates[mutType] * abs(property_dict[n_aa] - property_dict[aa])
                    else:
                        ermc += freq*mutRates[mutType] * max(0, property_dict[n_aa] - property_dict[aa])
                    
    return ermc
```

Number of nitrogen atoms in amino acids:

```python
# number of nitrogen atoms; Shenhav and Zeevi, Science, 2020
nitrogen_atoms = {'A': 1, 'C': 1, 'D': 1, 'E': 1, 'F': 1, 'G': 1, 'H': 3, 'I': 1, 'K': 2, 'L': 1, 'M':1, 'N': 2, 
    'P': 1, 'Q': 2, 'R': 4, 'S': 1, 'T': 1, 'V': 1, 'W': 2, 'Y': 1}
```

Generate 10,000 codes by amino acid permutation and for each compute ERMC using the ocean parameters and number of codons of nitrogen-rich amino acids:

```python
n_codes = 10000

# ocean mutation rates and codon abundances
mutRates = {}
with open("../data/mutabuns_ocean.csv", 'r') as f:
    data = f.read().split()[1:]
    for item in data:
        splitData = item.split(",")
        mutRates[splitData[0]] = float(splitData[1])
codons_frequencies = {}
with open("../data/codonabuns_ocean.csv", 'r') as cf:
    cf_lines = cf.read().split("\n")[1:]
    for item in cf_lines:
        splitItem = item.split(",")
        codons_frequencies[splitItem[0]] = float(splitItem[1])
        
# the standard code
standard_code = read_code("../data/code_standard.tsv")

# array to store the results
res = np.zeros(shape=(n_codes,2))

# generate the codes
for i in range(n_codes):
    code = rand_aa_permutation(standard_code, i)
    ermc_n = ERMC(code, nitrogen_atoms, mutRates, codons_frequencies)
    num_N_codons = number_N_rich_codons(code)
    res[i,:] = [ermc_n, num_N_codons]
```

Correlation between ERMC in nitrogen and number of codons of nitrogen-rich amino acids:

```python
scipy.stats.pearsonr(res[:,0], res[:,1])
```

## 2. Structural properites of the codes generated by quartet shuffling 

Functions to decide if codons of nitrogen-rich amino acids have a 'line', 'square' or a 'diagonal' arrangement:

```python
def hamming_distance(s1, s2):
    return sum([1 for i in range(len(s1)) if s1[i]!=s2[i]])

def issquare(codons):
    sq = True
    for i in codons:
        dsts = [hamming_distance(i,j) for j in codons]
        sq = sq and sorted(dsts) == [0,1,1,2]
    return sq

def isdiag(codons):
    diag = True
    for i in codons:
        dsts = [hamming_distance(i,j) for j in codons]
        diag = diag and sorted(dsts) == [0,2,2,2]
    return diag

def isline(codons):
    line = True
    for i in codons:
        dsts = [hamming_distance(i,j) for j in codons]
        line = line and sorted(dsts) == [0,1,1,1]
    return line
```

Function to generate a code by quartet shuffling and classify it as 'line', 'square', 'diagonal' or 'other':

```python
transitions = {'A':'G', 'G':'A', 'C':'T', 'T':'C'}

# standard_code ... dictionary containing the standard code; keys = codons, values = amino acids
def rand_quartet_shuffling(standard_code, seed = 0):
    np.random.seed(seed)
    code = standard_code.copy()
    bases = ['A', 'C', 'G', 'T']
    # all possible pairs of bases
    pairs = ['TA', 'TG', 'TT', 'TC', 'CT', 'CC', 'CA', 'CG', 'AT', 'AC', 'AA', 'AG', 'GT', 'GC', 'GA', 'GG']
    # permutation of the pairs to construct the alternative codon
    perm_pairs = pairs.copy()
    np.random.shuffle(perm_pairs)
    
    # the "opal" stop codon is separated by a transition from the "ambre" stop codon
    stop_pair_1 = perm_pairs[0]
    # the transition can happen in the first, or in the second position
    transition_pos = np.random.randint(2)
    if transition_pos==0:
        stop_pair_2 = transitions[stop_pair_1[0]]+stop_pair_1[1]
    else:
        stop_pair_2 = stop_pair_1[0]+transitions[stop_pair_1[1]]
    tmp_pos = perm_pairs.index(stop_pair_2)
    perm_pairs[tmp_pos] = perm_pairs[1]
    perm_pairs[1] = stop_pair_2
    # construct the new code
    for i in range(len(pairs)):   
        for b in bases:
            old_codon = pairs[i]+b
            new_codon = perm_pairs[i]+b
            code[new_codon] = standard_code[old_codon]
            
    # classify the code is "line", "square", "diagonal" or "other" arrangement with respect to N-rich aas
    code_config = "other"
    # get the relevant codons
    pos_N_rich = ["CA", "CG", "AA", "AG"]
    new_pos_N_rich = [perm_pairs[i] for i in range(len(pairs)) if pairs[i] in pos_N_rich]
    if issquare(new_pos_N_rich):
        code_config = "square"
    elif isdiag(new_pos_N_rich):
        code_config = "diag"
    elif isline(new_pos_N_rich):
        code_config = "line"
                                            
    return code, code_config
```

Generate 10,000 codes, for each compute ERMC using the ocean parameters:

```python
n_codes = 10000

# ocean mutation rates and codon abundances
mutRates = {}
with open("../data/mutabuns_ocean.csv", 'r') as f:
    data = f.read().split()[1:]
    for item in data:
        splitData = item.split(",")
        mutRates[splitData[0]] = float(splitData[1])
codons_frequencies = {}
with open("../data/codonabuns_ocean.csv", 'r') as cf:
    cf_lines = cf.read().split("\n")[1:]
    for item in cf_lines:
        splitItem = item.split(",")
        codons_frequencies[splitItem[0]] = float(splitItem[1])
        
# the standard code
standard_code = read_code("../data/code_standard.tsv")

# arrays to store the results
ermc_line = []
ermc_square = []
ermc_diag = []
ermc_other = []

# generate the codes
for i in range(n_codes):
    code, config = rand_quartet_shuffling(standard_code, i)
    ermc_n = ERMC(code, nitrogen_atoms, mutRates, codons_frequencies)
    if config == "line":
        ermc_line += [ermc_n]
    elif config == "square":
        ermc_square += [ermc_n]
    elif config == "diag":
        ermc_diag += [ermc_n]
    else:
        ermc_other += [ermc_n]
```

Boxplot of ERMC for different code arrangements:

```python
data = [ermc_line, ermc_square, ermc_other, ermc_diag]

fig = plt.figure(figsize =(10, 7))  
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data)
plt.show()
```
